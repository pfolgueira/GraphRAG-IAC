```python
from graphrag.graph.neo4j_manager import Neo4jManager
from graphrag.ingestion.text_processor import TextProcessor

neo4j_manager = Neo4jManager()
entities = [
    "Person",
    "Job Title",
    "Government Agency",
    "Organization",
    "Political Body",
    "City",
    "State",
    "Country",
    "Event",
    "Political Movement",
    "Facility",
    "Date",
    "Time",
    "Quantity",
    "Social Media Platform",
    "Document",
    "Media Outlet"
]
text_processor = TextProcessor(neo4j_manager, entity_types=entities)

with open("../data/trump.txt", "r", encoding="utf-8") as file:
    content = file.read()
    text_processor.process_document(content)
```
