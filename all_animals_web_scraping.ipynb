```python
import cloudscraper
from bs4 import BeautifulSoup

url = "https://a-z-animals.com/animals/"

scraper = cloudscraper.create_scraper()
response = scraper.get(url)

print(response.status_code)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Comprobar si se ha cargado bien el html de la página
soup.title
```

```python
import string

alphabet = list(string.ascii_lowercase)
```

```python
id = f"h-animals-that-start-with-{alphabet[0]}"
id
```

```python
header = soup.select_one("h3#" + id)
header.get("id")
```

```python
container = header.find_next_sibling()

lista = container.find("ul")
items = lista.select("li > a")

for i in items:
    print(i.get("href"))
```

```python
animals_urls = []

for letter in alphabet:
    id = f"h-animals-that-start-with-{letter}"
    header = soup.select_one("h3#" + id)
    container = header.find_next_sibling()
    lista = container.find("ul")
    items = lista.select("li > a")

    for i in items:
        animals_urls.append(i.get("href"))
```
